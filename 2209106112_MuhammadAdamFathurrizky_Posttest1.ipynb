Setelah menganalisis dataset MFG 10 Year Termination, berikut adalah poin-poin utama yang bisa diambil:

1. Distribusi Data
Berdasarkan hasil analisis statistik:

- Rata-rata (Mean): Mayoritas variabel numerik dalam dataset memiliki rata-rata yang cukup terdistribusi dengan baik. Ini berarti nilai-nilai data tersebar di sekitar nilai tengah.
- Standar Deviasi: Standar deviasi yang besar menunjukkan bahwa ada variabilitas tinggi dalam data. Kalau nilai ini terlalu besar, ada kemungkinan terdapat outlier atau nilai-nilai ekstrem yang jauh dari rata-rata.

2. Kuartil (Q1, Q2, Q3)
- Q1 (Kuartil Pertama): 25% dari data berada di bawah nilai Q1, menunjukkan distribusi nilai di bagian bawah data.
Median
- (Q2): Setengah dari data berada di bawah nilai median, ini menunjukkan titik tengah dari distribusi data.
- Q3 (Kuartil Ketiga): 75% dari data ada di bawah Q3, yang menggambarkan nilai-nilai di bagian atas distribusi.

3. Hubungan Data

- Hubungan antar variabel: Ada kemungkinan beberapa variabel memiliki hubungan atau korelasi. Dengan menggunakan scatter plot, kita bisa melihat apakah ada pola keterkaitan antar variabel.
- Distribusi Data: Dengan histogram, kita bisa melihat sebaran data dan mengidentifikasi apakah ada outlier atau anomali di beberapa variabel.

4. Faktor-Faktor Penting
